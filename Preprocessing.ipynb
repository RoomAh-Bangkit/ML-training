# EDA and Balancing Data

## Inspect

```python
# @title Hiding the code lol

root_dir = ''
# root_dir = 'BACKUP/'
train_dir = f'./{root_dir}House_Room_Dataset/Training'
validation_dir = f'./{root_dir}House_Room_Dataset/Validation'
testing_dir = f'./{root_dir}House_Room_Dataset/Testing'
archive_dir = f'./{root_dir}House_Room_Dataset/Archive'

rooms = ['Bathroom','Bedroom','Dinning','Kitchen','Livingroom']

count_train = []
count_validation = []
count_testing = []
count_archive = []

for room in rooms:
    n = len(os.listdir(f"{train_dir}/{room}"))
    count_train.append(n)
    
    n = len(os.listdir(f"{validation_dir}/{room}"))
    count_validation.append(n)

    n = len(os.listdir(f"{testing_dir}/{room}"))
    count_testing.append(n)

    n = len(os.listdir(f"{archive_dir}/{room}"))
    count_archive.append(n)

sum_train = sum(count_train)
perc_train = [round(count/sum_train*100,2) for count in count_train]

sum_validation = sum(count_validation)
perc_validation = [round(count/sum_validation*100,2) for count in count_validation]

deltas = []
# max_* = how many photos for *
max_train = [1600 for i in range(5)]
max_test = [100 for i in range(5)]

add_train = [_max_train-count_train for (count_train,_max_train,count_test) in zip(count_train,max_train,count_testing)]
add_test = [_max_test-count_test for (count_train,_max_test,count_test) in zip(count_train,max_test,count_testing)]
less_val = [_add_train+_add_test for (_add_train,_add_test) in zip(add_train,add_test)]
left_val = [count_val-_less_val for (count_val,_less_val) in zip(count_validation,less_val)]

print("Train")
mdf_train = pd.DataFrame(np.array([rooms,perc_train,count_train,add_train,max_train]).T,
                         columns=['Name','Perc','Count','Add','Sum'])
mdf_train.set_index('Name', inplace=True)
print(mdf_train, end='\n\n\n')

print("Validation")
mdf_validation = pd.DataFrame(np.array([rooms,perc_validation,count_validation,less_val,left_val]).T,
                              columns=['Name','Perc','Count','Less','Left'])
mdf_validation.set_index('Name', inplace=True)
print(mdf_validation, end='\n\n\n')

print("Testing")
mdf_testing = pd.DataFrame(np.array([rooms,count_testing,add_test,max_test]).T,
                              columns=['Name','Count','Add','Sum'])
mdf_testing.set_index('Name', inplace=True)
print(mdf_testing, end='\n\n\n')

print("Archive")
mdf_archive = pd.DataFrame(np.array([rooms,count_archive]).T,
                              columns=['Name','Count'])
mdf_archive.set_index('Name', inplace=True)
print(mdf_archive, end='\n\n\n')
```
